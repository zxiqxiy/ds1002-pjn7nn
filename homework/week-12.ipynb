## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt

slices = [56, 66, 24, 124, 47]
labels = ["US", "UK", "MX", "CA", "JP"]

plt.pie(slices, labels=labels, autopct='%1.2f%%', startangle=90)

plt.title("Distribution of Values")

plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)

plt.hist(n, bins=30, color='pink', edgecolor='black')

plt.title('Histogram of Random Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.polynomial import Polynomial

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

df = pd.DataFrame({'x': x, 'y': y})

plt.scatter(df['x'], df['y'], label='Data Points')

coeffs = np.polyfit(df['x'], df['y'], 1)
trendline = Polynomial(coeffs)
x_values = np.linspace(df['x'].min(), df['x'].max(), 100)
y_values = trendline(x_values)

plt.plot(x_values, y_values, color='purple', label='Trendline (degree=1)')

plt.title('Scatterplot with Trendline')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.legend()

plt.show()
```
